```python
import os
import pandas as pd

# -------------------------------
# Step 1: Set your folder path
# -------------------------------
folder_path = r"C:\Users\Admin\Desktop\data ar\Data science\1 CR\project\anyisis\YouTube Trending\Data"

# -------------------------------
# Step 2: Automatically find the CSV file
# -------------------------------
csv_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]

if not csv_files:
    print("No CSV file found in the folder!")
    exit()
elif len(csv_files) > 1:
    print("Multiple CSV files found. Using the first one found:", csv_files[0])

csv_file_name = csv_files[0]
full_path = os.path.join(folder_path, csv_file_name)
print("CSV file found:", csv_file_name)

# -------------------------------
# Step 3: Load the CSV
# -------------------------------
try:
    df = pd.read_csv(full_path)
    print(f"CSV loaded successfully! Total rows: {len(df)}")
except Exception as e:
    print("Error loading CSV:", e)
    exit()

# -------------------------------
# Step 4: Split CSV into smaller chunks
# -------------------------------
chunk_size = 50000  # number of rows per chunk
num_chunks = (len(df) // chunk_size) + 1

for i, start in enumerate(range(0, len(df), chunk_size), 1):
    df_chunk = df[start:start + chunk_size]
    chunk_file_name = f"YouTube_Trending_2026_Part{i}.csv"
    chunk_path = os.path.join(folder_path, chunk_file_name)
    df_chunk.to_csv(chunk_path, index=False)
    print(f"Saved chunk {i} with {len(df_chunk)} rows -> {chunk_file_name}")

# -------------------------------
# Step 5: Done message
# -------------------------------
print(f"\nAll done! {num_chunks} chunks created in:\n{folder_path}")
```
